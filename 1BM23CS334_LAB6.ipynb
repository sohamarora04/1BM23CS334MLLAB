```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

iris_df = pd.read_csv('iris.csv')
X = iris_df.drop('species', axis=1)
y = iris_df['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

iris_clf = DecisionTreeClassifier(random_state=42)
iris_clf.fit(X_train, y_train)

y_pred = iris_clf.predict(X_test)

print("--- Iris Dataset Performance ---")
print(f"Accuracy Score: {accuracy_score(y_test, y_pred):.4f}")
print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
display(pd.DataFrame(cm, index=iris_clf.classes_, columns=iris_clf.classes_))
```

```python
drug_df = pd.read_csv('drug.csv')
drug_df_encoded = pd.get_dummies(drug_df.drop('Drug', axis=1))
X_drug = drug_df_encoded
y_drug = drug_df['Drug']

X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(X_drug, y_drug, test_size=0.20, random_state=42)

drug_clf = DecisionTreeClassifier(random_state=42)
drug_clf.fit(X_train_d, y_train_d)

y_pred_d = drug_clf.predict(X_test_d)

print("--- Drug Dataset Performance ---")
print(f"Accuracy Score: {accuracy_score(y_test_d, y_pred_d):.4f}")
print("\nConfusion Matrix:")
cm_d = confusion_matrix(y_test_d, y_pred_d)
display(pd.DataFrame(cm_d, index=drug_clf.classes_, columns=drug_clf.classes_))
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

petrol_df = pd.read_csv('petrol_consumption.csv')
X_p = petrol_df.drop('Petrol_Consumption', axis=1)
y_p = petrol_df['Petrol_Consumption']

X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(X_p, y_p, test_size=0.20, random_state=42)

regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(X_train_p, y_train_p)

y_pred_p = regressor.predict(X_test_p)

mae = mean_absolute_error(y_test_p, y_pred_p)
mse = mean_squared_error(y_test_p, y_pred_p)
rmse = np.sqrt(mse)

print("--- Petrol Consumption Regression Performance ---")
print(f"Mean Absolute Error: {mae:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"Root Mean Squared Error: {rmse:.4f}")

feat_importances = pd.Series(regressor.feature_importances_, index=X_p.columns)
print("\nFeature Importances:")
print(feat_importances.sort_values(ascending=False))
```
